# Digit DP

Digit DP counts numbers with a property without listing them all.

This notebook explains the idea in plain language first, then gives Java and Python templates you can adapt in LeetCode and Codeforces.

## When to use it

Use it for counts under N with digit sum, repeated digits, divisibility, or pattern restrictions.

## Layman idea

Build the number digit by digit with a tight flag that respects the upper bound.

## Java implementation

```java
import java.util.*;
class DigitDP { char[] s; Map<String,Long> memo; long count(long n,int limit){ s=Long.toString(n).toCharArray(); memo=new HashMap<>(); return dfs(0,0,true,limit); } long dfs(int pos,int sum,boolean tight,int limit){ if(sum>limit) return 0; if(pos==s.length) return 1; String key=pos+","+sum; if(!tight && memo.containsKey(key)) return memo.get(key); int max=tight?s[pos]-'0':9; long ans=0; for(int d=0;d<=max;d++) ans+=dfs(pos+1,sum+d,tight&&d==max,limit); if(!tight) memo.put(key,ans); return ans; } }
```

## Python implementation

```python
from functools import lru_cache
def count_digit_sum_at_most(n, limit):
    digits=list(map(int,str(n)))
    @lru_cache(None)
    def dfs(pos,total,tight):
        if total > limit: return 0
        if pos == len(digits): return 1
        mx = digits[pos] if tight else 9
        return sum(dfs(pos+1, total+d, tight and d==mx) for d in range(mx+1))
    return dfs(0,0,True)
```

## Complexity

O(digits * states * 10), space is memo states.

## Common mistake

Do not memoize tight states as if every prefix were the same.

## Practice checklist

- Write one tiny example by hand.
