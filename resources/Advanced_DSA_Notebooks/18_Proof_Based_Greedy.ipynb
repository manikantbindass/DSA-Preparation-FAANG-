# Proof-Based Greedy

Greedy works only when the local best choice can be proven safe.

This notebook explains the idea in plain language first, then gives Java and Python templates you can adapt in LeetCode and Codeforces.

## When to use it

Use it when sorting by earliest end, smallest cost, largest gain, or deadline looks natural.

## Layman idea

Exchange argument: replace the first optimal choice with the greedy choice and show nothing gets worse.

## Java implementation

```java
import java.util.*;
class ProofGreedy { static int eraseOverlap(int[][] in){ Arrays.sort(in,(a,b)->a[1]-b[1]); int keep=0,last=Integer.MIN_VALUE; for(int[] x:in) if(x[0]>=last){keep++; last=x[1];} return in.length-keep; } }
```

## Python implementation

```python
def erase_overlap(intervals):
    intervals.sort(key=lambda x: x[1])
    keep = 0; last = -10**30
    for start, end in intervals:
        if start >= last: keep += 1; last = end
    return len(intervals) - keep
```

## Complexity

Usually O(n log n) because sorting dominates.

## Common mistake

If you cannot prove it, consider DP instead.
