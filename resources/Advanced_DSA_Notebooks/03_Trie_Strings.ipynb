# Trie - Strings

A trie stores words letter by letter. Words with the same prefix share the same path.

This notebook explains the idea in plain language first, then gives Java and Python templates you can adapt in LeetCode and Codeforces.

## When to use it

Use it for prefix search, dictionaries, autocomplete, word break, and bitwise XOR tries.

## Layman idea

Walk from root through characters. Create missing nodes while inserting; only follow existing nodes while searching.

## Java implementation

```java
class Trie {
  static class Node { Node[] next = new Node[26]; boolean word; }
  Node root = new Node();
  void insert(String s){ Node cur=root; for(char c:s.toCharArray()){ int i=c-'a'; if(cur.next[i]==null) cur.next[i]=new Node(); cur=cur.next[i]; } cur.word=true; }
  boolean search(String s){ Node x=walk(s); return x!=null && x.word; }
  boolean startsWith(String s){ return walk(s)!=null; }
  Node walk(String s){ Node cur=root; for(char c:s.toCharArray()){ int i=c-'a'; if(cur.next[i]==null) return null; cur=cur.next[i]; } return cur; }
}
```

## Python implementation

```python
class Node:
    def __init__(self): self.children = {}; self.word = False
class Trie:
    def __init__(self): self.root = Node()
    def insert(self, s):
        cur = self.root
        for ch in s: cur = cur.children.setdefault(ch, Node())
        cur.word = True
    def walk(self, s):
        cur = self.root
        for ch in s:
            if ch not in cur.children: return None
            cur = cur.children[ch]
        return cur
    def search(self, s):
        node = self.walk(s); return node is not None and node.word
    def starts_with(self, s): return self.walk(s) is not None
```

## Complexity

Insert/search O(length), space O(total characters).

## Common mistake

Prefix existing is not the same as word existing; keep an end marker.
