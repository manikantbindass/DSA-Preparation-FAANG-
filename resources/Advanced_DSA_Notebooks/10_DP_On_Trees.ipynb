# DP on Trees

Tree DP lets each node return useful information about its subtree.

This notebook explains the idea in plain language first, then gives Java and Python templates you can adapt in LeetCode and Codeforces.

## When to use it

Use it for subtree choices, diameter, independent set, rerooting, and path answers.

## Layman idea

Root the tree, solve children first, combine their answers at the parent.

## Java implementation

```java
import java.util.*;
class TreeDP { List<List<Integer>> g; int[] val; long[] dfs(int u,int p){ long take=val[u], skip=0; for(int v:g.get(u)) if(v!=p){ long[] c=dfs(v,u); take+=c[0]; skip+=Math.max(c[0],c[1]); } return new long[]{skip,take}; } long solve(List<List<Integer>> g,int[] val){ this.g=g; this.val=val; long[] a=dfs(0,-1); return Math.max(a[0],a[1]); }
}
```

## Python implementation

```python
def tree_independent_set(g, val):
    def dfs(u, p):
        take, skip = val[u], 0
        for v in g[u]:
            if v != p:
                child_skip, child_take = dfs(v, u)
                take += child_skip; skip += max(child_skip, child_take)
        return skip, take
    return max(dfs(0, -1))
```

## Complexity

Usually O(n) time and O(n) stack/DP space.

## Common mistake

Pass parent or visited to avoid walking back forever.
