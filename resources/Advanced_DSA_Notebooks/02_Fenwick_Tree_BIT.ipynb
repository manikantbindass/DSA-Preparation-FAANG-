# Fenwick Tree - Binary Indexed Tree

A Fenwick tree is a smaller cousin of segment tree for prefix sums. Each index stores a basket of responsibility.

This notebook explains the idea in plain language first, then gives Java and Python templates you can adapt in LeetCode and Codeforces.

## When to use it

Use it for point update plus prefix/range sum, inversion count, frequencies, and order queries.

## Layman idea

The lowest set bit tells how big a basket is. Update jumps upward; prefix query jumps downward.

## Java implementation

```java
class Fenwick {
  long[] bit; Fenwick(int n){ bit=new long[n+1]; }
  void add(int i,long delta){ for(i++; i<bit.length; i+=i&-i) bit[i]+=delta; }
  long sum(int i){ long s=0; for(i++; i>0; i-=i&-i) s+=bit[i]; return s; }
  long range(int l,int r){ return sum(r) - (l==0 ? 0 : sum(l-1)); }
}
```

## Python implementation

```python
class Fenwick:
    def __init__(self, n): self.bit = [0] * (n + 1)
    def add(self, i, delta):
        i += 1
        while i < len(self.bit): self.bit[i] += delta; i += i & -i
    def sum(self, i):
        i += 1; ans = 0
        while i > 0: ans += self.bit[i]; i -= i & -i
        return ans
    def range(self, l, r): return self.sum(r) - (0 if l == 0 else self.sum(l - 1))
```

## Complexity

Update/query O(log n), space O(n).

## Common mistake

Internally use 1-based indexing, even if the input is 0-based.
