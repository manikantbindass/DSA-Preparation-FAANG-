# Bitmask DP

Bitmask DP stores a chosen set inside bits of an integer.

This notebook explains the idea in plain language first, then gives Java and Python templates you can adapt in LeetCode and Codeforces.

## When to use it

Use it when n is small and state depends on which items are already used.

## Layman idea

mask says what is chosen. Add item i by moving to mask | (1 << i).

## Java implementation

```java
import java.util.*;
class BitmaskDP { static int assign(int[][] cost){ int n=cost.length, full=1<<n; int[] dp=new int[full]; Arrays.fill(dp,1_000_000_000); dp[0]=0; for(int mask=0;mask<full;mask++){ int worker=Integer.bitCount(mask); for(int job=0;job<n && worker<n;job++) if((mask&(1<<job))==0) dp[mask|1<<job]=Math.min(dp[mask|1<<job], dp[mask]+cost[worker][job]); } return dp[full-1]; } }
```

## Python implementation

```python
def assignment_cost(cost):
    n=len(cost); full=1<<n; dp=[10**18]*full; dp[0]=0
    for mask in range(full):
        worker = mask.bit_count()
        if worker == n: continue
        for job in range(n):
            if not (mask & (1<<job)): dp[mask | (1<<job)] = min(dp[mask | (1<<job)], dp[mask] + cost[worker][job])
    return dp[-1]
```

## Complexity

Often O(n * 2^n) time and O(2^n) space.

## Common mistake

Check constraints first; bitmask DP grows very fast.

## Practice checklist
