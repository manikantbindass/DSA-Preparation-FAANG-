# Bellman-Ford

Bellman-Ford finds shortest paths even with negative edges, and can detect reachable negative cycles.

This notebook explains the idea in plain language first, then gives Java and Python templates you can adapt in LeetCode and Codeforces.

## When to use it

Use it when weights can be negative or a negative cycle matters.

## Layman idea

Relax all edges n-1 times. If a further relaxation works, a negative cycle exists.

## Java implementation

```java
import java.util.*;
class BellmanFord {
  static long INF = Long.MAX_VALUE / 4;
  static long[] shortest(int n,int[][] edges,int src){ long[] d=new long[n]; Arrays.fill(d, INF); d[src]=0; for(int i=0;i<n-1;i++){ boolean changed=false; for(int[] e:edges){ if(d[e[0]]!=INF && d[e[0]]+e[2]<d[e[1]]){ d[e[1]]=d[e[0]]+e[2]; changed=true; } } if(!changed) break; } return d; }
}
```

## Python implementation

```python
def bellman_ford(n, edges, src):
    INF = 10**30; dist = [INF] * n; dist[src] = 0
    for _ in range(n - 1):
        changed = False
        for u, v, w in edges:
            if dist[u] != INF and dist[u] + w < dist[v]: dist[v] = dist[u] + w; changed = True
        if not changed: break
    neg_cycle = any(dist[u] != INF and dist[u] + w < dist[v] for u, v, w in edges)
    return dist, neg_cycle
```

## Complexity

O(VE) time, O(V) space.

## Common mistake

Never relax from INF; it creates fake distances.
