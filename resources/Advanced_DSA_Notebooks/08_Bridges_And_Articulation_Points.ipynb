# Bridges and Articulation Points

Bridges and articulation points are weak spots whose removal disconnects a graph.

This notebook explains the idea in plain language first, then gives Java and Python templates you can adapt in LeetCode and Codeforces.

## When to use it

Use it for critical connections, network reliability, and graph cut problems.

## Layman idea

DFS discovery time and low-link values show whether a subtree can climb back above its parent.

## Java implementation

```java
import java.util.*;
class Bridges { int timer=0; List<int[]> ans=new ArrayList<>(); int[] tin,low;
  List<int[]> find(List<List<Integer>> g){ int n=g.size(); tin=new int[n]; low=new int[n]; Arrays.fill(tin,-1); for(int i=0;i<n;i++) if(tin[i]==-1) dfs(i,-1,g); return ans; }
  void dfs(int u,int p,List<List<Integer>> g){ tin[u]=low[u]=timer++; for(int v:g.get(u)){ if(v==p) continue; if(tin[v]!=-1) low[u]=Math.min(low[u],tin[v]); else { dfs(v,u,g); low[u]=Math.min(low[u],low[v]); if(low[v]>tin[u]) ans.add(new int[]{u,v}); } } }
}
```

## Python implementation

```python
def bridges(g):
    n=len(g); tin=[-1]*n; low=[0]*n; timer=0; ans=[]
    def dfs(u,p):
        nonlocal timer; tin[u]=low[u]=timer; timer+=1
        for v in g[u]:
            if v == p: continue
            if tin[v] != -1: low[u] = min(low[u], tin[v])
            else:
                dfs(v,u); low[u] = min(low[u], low[v])
                if low[v] > tin[u]: ans.append((u,v))
    for i in range(n):
        if tin[i] == -1: dfs(i,-1)
    return ans
```

## Complexity

O(V+E) time, O(V) extra space.

## Common mistake

For multiedges, skip by edge id instead of only parent node.
