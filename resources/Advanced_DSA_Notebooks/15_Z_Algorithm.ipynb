# Z Algorithm

The Z array tells how much each suffix matches the prefix.

This notebook explains the idea in plain language first, then gives Java and Python templates you can adapt in LeetCode and Codeforces.

## When to use it

Use it for pattern matching and prefix-similarity problems.

## Layman idea

Keep a matching window [l,r] and reuse known matches when the next index is inside it.

## Java implementation

```java
class ZAlgo { static int[] z(String s){ int n=s.length(), l=0, r=0; int[] z=new int[n]; for(int i=1;i<n;i++){ if(i<=r) z[i]=Math.min(r-i+1,z[i-l]); while(i+z[i]<n && s.charAt(z[i])==s.charAt(i+z[i])) z[i]++; if(i+z[i]-1>r){l=i;r=i+z[i]-1;} } return z; } }
```

## Python implementation

```python
def z_function(s):
    n=len(s); z=[0]*n; left=right=0
    for i in range(1,n):
        if i <= right: z[i] = min(right-i+1, z[i-left])
        while i+z[i] < n and s[z[i]] == s[i+z[i]]: z[i] += 1
        if i+z[i]-1 > right: left, right = i, i+z[i]-1
    return z
```
