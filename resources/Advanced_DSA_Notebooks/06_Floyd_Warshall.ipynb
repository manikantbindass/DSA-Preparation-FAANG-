# Floyd-Warshall

Floyd-Warshall finds shortest paths between every pair of nodes.

This notebook explains the idea in plain language first, then gives Java and Python templates you can adapt in LeetCode and Codeforces.

## When to use it

Use it when n is small and all-pairs distances are needed.

## Layman idea

Try every node k as a middle stop between i and j.

## Java implementation

```java
class FloydWarshall {
  static long INF = Long.MAX_VALUE / 4;
  static void run(long[][] d){ int n=d.length; for(int k=0;k<n;k++) for(int i=0;i<n;i++) for(int j=0;j<n;j++) if(d[i][k]!=INF && d[k][j]!=INF) d[i][j]=Math.min(d[i][j], d[i][k]+d[k][j]); }
}
```

## Python implementation

```python
def floyd_warshall(d):
    INF = 10**30; n = len(d)
    for k in range(n):
        for i in range(n):
            for j in range(n):
                if d[i][k] != INF and d[k][j] != INF: d[i][j] = min(d[i][j], d[i][k] + d[k][j])
    return d
```

## Complexity

O(n^3) time, O(n^2) space.

## Common mistake

Set dist[i][i] = 0 and keep the smallest duplicate edge.
