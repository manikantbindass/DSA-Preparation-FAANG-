# Strongly Connected Components - Kosaraju

An SCC is a directed group where every node can reach every other node.

This notebook explains the idea in plain language first, then gives Java and Python templates you can adapt in LeetCode and Codeforces.

## When to use it

Use it for condensation DAGs, dependency cycles, and 2-SAT foundations.

## Layman idea

DFS finish order first, reverse the graph, then peel SCCs in reverse finish order.

## Java implementation

```java
import java.util.*;
class Kosaraju {
  List<List<Integer>> scc(List<List<Integer>> g){ int n=g.size(); List<List<Integer>> r=new ArrayList<>(); for(int i=0;i<n;i++) r.add(new ArrayList<>()); for(int u=0;u<n;u++) for(int v:g.get(u)) r.get(v).add(u); boolean[] seen=new boolean[n]; List<Integer> order=new ArrayList<>(); for(int i=0;i<n;i++) if(!seen[i]) dfs1(i,g,seen,order); Arrays.fill(seen,false); Collections.reverse(order); List<List<Integer>> comps=new ArrayList<>(); for(int x:order) if(!seen[x]){ List<Integer> c=new ArrayList<>(); dfs2(x,r,seen,c); comps.add(c);} return comps; }
  void dfs1(int u,List<List<Integer>> g,boolean[] s,List<Integer> o){ s[u]=true; for(int v:g.get(u)) if(!s[v]) dfs1(v,g,s,o); o.add(u); }
  void dfs2(int u,List<List<Integer>> g,boolean[] s,List<Integer> c){ s[u]=true; c.add(u); for(int v:g.get(u)) if(!s[v]) dfs2(v,g,s,c); }
}
```

## Python implementation

```python
def kosaraju(g):
    n=len(g); rev=[[] for _ in range(n)]
    for u in range(n):
        for v in g[u]: rev[v].append(u)
    seen=[False]*n; order=[]
    def dfs1(u):
        seen[u]=True
        for v in g[u]:
            if not seen[v]: dfs1(v)
        order.append(u)
    def dfs2(u, comp):
        seen[u]=True; comp.append(u)
        for v in rev[u]:
            if not seen[v]: dfs2(v, comp)
    for i in range(n):
        if not seen[i]: dfs1(i)
    seen=[False]*n; comps=[]
    for x in reversed(order):
        if not seen[x]: comp=[]; dfs2(x, comp); comps.append(comp)
    return comps
```

## Complexity

O(V+E) time, O(V+E) space.

## Common mistake

SCC is for directed graphs; undirected graphs just need connected components.
