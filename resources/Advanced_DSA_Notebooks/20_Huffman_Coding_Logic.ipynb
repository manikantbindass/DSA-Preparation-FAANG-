# Huffman Coding Logic

Huffman coding gives frequent items short codes by repeatedly merging the two smallest frequencies.

This notebook explains the idea in plain language first, then gives Java and Python templates you can adapt in LeetCode and Codeforces.

## When to use it

Use it for compression logic, optimal merge pattern, and heap-based greedy reasoning.

## Layman idea

A min-heap always exposes the two cheapest pieces to merge next.

## Java implementation

```java
import java.util.*;
class Huffman { static int cost(int[] freq){ PriorityQueue<Integer> pq=new PriorityQueue<>(); for(int f:freq) if(f>0) pq.offer(f); int ans=0; while(pq.size()>1){ int a=pq.poll(), b=pq.poll(); ans+=a+b; pq.offer(a+b);} return ans; } }
```

## Python implementation

```python
import heapq
def huffman_cost(freq):
    heap=[x for x in freq if x > 0]; heapq.heapify(heap); ans=0
    while len(heap) > 1:
        a=heapq.heappop(heap); b=heapq.heappop(heap); ans += a+b; heapq.heappush(heap, a+b)
    return ans
```

## Complexity

O(n log n) time, O(n) space.

## Common mistake

Codes must be prefix-free; the merge tree guarantees that.
