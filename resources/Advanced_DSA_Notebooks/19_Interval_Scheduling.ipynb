# Interval Scheduling

Interval scheduling chooses the most non-overlapping intervals.

This notebook explains the idea in plain language first, then gives Java and Python templates you can adapt in LeetCode and Codeforces.

## When to use it

Use it for meetings, jobs, activities, and selecting maximum compatible ranges.

## Layman idea

Sort by end time and keep every interval that starts after the last chosen end.

## Java implementation

```java
import java.util.*;
class ProofGreedy { static int eraseOverlap(int[][] in){ Arrays.sort(in,(a,b)->a[1]-b[1]); int keep=0,last=Integer.MIN_VALUE; for(int[] x:in) if(x[0]>=last){keep++; last=x[1];} return in.length-keep; } }
```

## Python implementation

```python
def erase_overlap(intervals):
    intervals.sort(key=lambda x: x[1])
    keep = 0; last = -10**30
    for start, end in intervals:
        if start >= last: keep += 1; last = end
    return len(intervals) - keep
```

## Complexity

O(n log n) time, O(1) extra space after sorting.

## Common mistake

Know whether touching endpoints count as overlapping.
