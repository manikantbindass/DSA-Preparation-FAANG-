# Rabin-Karp - Rolling Hash

Rabin-Karp compares substrings by numeric fingerprints.

This notebook explains the idea in plain language first, then gives Java and Python templates you can adapt in LeetCode and Codeforces.

## When to use it

Use it for many substring comparisons, duplicate substring, and multi-pattern checks.

## Layman idea

Prefix hashes let you get any substring hash in O(1).

## Java implementation

```java
class RollingHash { long MOD=1_000_000_007L, BASE=911382323L; long[] h,p; RollingHash(String s){ int n=s.length(); h=new long[n+1]; p=new long[n+1]; p[0]=1; for(int i=0;i<n;i++){ h[i+1]=(h[i]*BASE+s.charAt(i))%MOD; p[i+1]=p[i]*BASE%MOD; } } long get(int l,int r){ long x=(h[r+1]-h[l]*p[r-l+1])%MOD; return x<0?x+MOD:x; } }
```

## Python implementation

```python
class RollingHash:
    MOD=1_000_000_007; BASE=911_382_323
    def __init__(self, s):
        n=len(s); self.h=[0]*(n+1); self.p=[1]*(n+1)
        for i,ch in enumerate(s): self.h[i+1]=(self.h[i]*self.BASE+ord(ch))%self.MOD; self.p[i+1]=self.p[i]*self.BASE%self.MOD
    def get(self, l, r): return (self.h[r+1] - self.h[l] * self.p[r-l+1]) % self.MOD
```

## Complexity

Preprocess O(n), query O(1), space O(n).

## Common mistake

Hashes can collide; double hash or verify when needed.
