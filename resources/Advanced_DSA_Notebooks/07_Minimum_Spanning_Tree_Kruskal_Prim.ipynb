# Minimum Spanning Tree - Kruskal and Prim

An MST connects every node with minimum total cost and no cycles.

This notebook explains the idea in plain language first, then gives Java and Python templates you can adapt in LeetCode and Codeforces.

## When to use it

Use it for cheapest network connection problems on undirected graphs.

## Layman idea

Kruskal sorts edges and uses DSU. Prim grows a tree with a priority queue.

## Java implementation

```java
import java.util.*;
class Kruskal {
  static int mst(int n,int[][] edges){ Arrays.sort(edges,(a,b)->a[2]-b[2]); DSU d=new DSU(n); int cost=0,used=0; for(int[] e:edges) if(d.union(e[0],e[1])){ cost+=e[2]; if(++used==n-1) break; } return used==n-1?cost:-1; }
  static class DSU{ int[] p,s; DSU(int n){p=new int[n];s=new int[n];for(int i=0;i<n;i++){p[i]=i;s[i]=1;}} int find(int x){return p[x]==x?x:(p[x]=find(p[x]));} boolean union(int a,int b){int x=find(a),y=find(b); if(x==y)return false; if(s[x]<s[y]){int t=x;x=y;y=t;} p[y]=x;s[x]+=s[y];return true;} }
}
```

## Python implementation

```python
class DSU:
    def __init__(self, n): self.p = list(range(n)); self.s = [1] * n
    def find(self, x):
        if self.p[x] != x: self.p[x] = self.find(self.p[x])
        return self.p[x]
    def union(self, a, b):
        a, b = self.find(a), self.find(b)
        if a == b: return False
        if self.s[a] < self.s[b]: a, b = b, a
        self.p[b] = a; self.s[a] += self.s[b]; return True
def kruskal(n, edges):
    d = DSU(n); cost = used = 0
    for u, v, w in sorted(edges, key=lambda x: x[2]):
        if d.union(u, v): cost += w; used += 1
    return cost if used == n - 1 else -1
```

## Complexity

Kruskal O(E log E); Prim O(E log V).

## Common mistake

Disconnected graphs have a spanning forest, not one MST.
