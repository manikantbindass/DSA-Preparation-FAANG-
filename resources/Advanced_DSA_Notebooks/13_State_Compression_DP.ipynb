# State Compression DP

State compression packs a large state into a small code, usually a bitmask.

This notebook explains the idea in plain language first, then gives Java and Python templates you can adapt in LeetCode and Codeforces.

## When to use it

Use it for row-by-row grid DP, tiling, and profile DP.

## Layman idea

Keep only the previous compressed state and test compatibility with the current state.

## Java implementation

```java
import java.util.*;
class StateCompression { static long ways(int rows,int cols){ List<Integer> valid=new ArrayList<>(); for(int m=0;m<(1<<cols);m++) if((m&(m<<1))==0) valid.add(m); Map<Integer,Long> dp=new HashMap<>(); for(int m:valid) dp.put(m,1L); for(int r=1;r<rows;r++){ Map<Integer,Long> next=new HashMap<>(); for(int cur:valid){ long w=0; for(int prev:valid) if((cur&prev)==0) w+=dp.get(prev); next.put(cur,w);} dp=next;} long ans=0; for(long x:dp.values()) ans+=x; return ans;} }
```

## Python implementation

```python
def compressed_grid_ways(rows, cols):
    valid=[m for m in range(1<<cols) if (m & (m<<1)) == 0]
    dp={m:1 for m in valid}
    for _ in range(1, rows):
        dp={cur:sum(dp[prev] for prev in valid if not (cur & prev)) for cur in valid}
    return sum(dp.values())
```

## Complexity

Depends on states; row masks are often O(rows * states^2).

## Common mistake

Write compatibility rules carefully and test tiny grids.

## Practice checklist

- Identify the state or data structure invariant.
- Write one tiny example by hand.
- Check edge cases: empty input, one node/item, duplicate values, and disconnected graphs where relevant.
