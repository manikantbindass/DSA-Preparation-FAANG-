# KMP Algorithm

KMP searches a pattern without rechecking characters unnecessarily.

This notebook explains the idea in plain language first, then gives Java and Python templates you can adapt in LeetCode and Codeforces.

## When to use it

Use it for exact pattern search, prefix-suffix tasks, and repeated substring problems.

## Layman idea

The LPS array says where the pattern pointer should jump after mismatch.

## Java implementation

```java
import java.util.*;
class KMP { static int[] lps(String p){ int[] a=new int[p.length()]; for(int i=1,len=0;i<p.length();){ if(p.charAt(i)==p.charAt(len)) a[i++]=++len; else if(len>0) len=a[len-1]; else i++; } return a; } static List<Integer> search(String s,String p){ int[] l=lps(p); List<Integer> ans=new ArrayList<>(); for(int i=0,j=0;i<s.length();){ if(s.charAt(i)==p.charAt(j)){i++;j++;} if(j==p.length()){ans.add(i-j);j=l[j-1];} else if(i<s.length()&&s.charAt(i)!=p.charAt(j)){ if(j>0) j=l[j-1]; else i++; } } return ans; } }
```

## Python implementation

```python
def lps(p):
    a=[0]*len(p); length=0; i=1
    while i < len(p):
        if p[i] == p[length]: length += 1; a[i] = length; i += 1
        elif length: length = a[length-1]
        else: i += 1
    return a
def kmp(s, p):
    l=lps(p); ans=[]; i=j=0
    while i < len(s):
        if s[i] == p[j]: i += 1; j += 1
        if j == len(p): ans.append(i-j); j = l[j-1]
        elif i < len(s) and s[i] != p[j]:
            if j: j = l[j-1]
            else: i += 1
    return ans
```

## Complexity

O(n+m) time, O(m) space.

## Common mistake

LPS belongs to the pattern, not the text.
