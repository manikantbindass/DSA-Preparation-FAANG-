# Computational Geometry

## 1. Orientation
Orientation of three ordered points (P, Q, R) can be:
- Counter-clockwise
- Clockwise
- Collinear
Determined by the sign of: (y2 - y1)(x3 - x2) - (y3 - y2)(x2 - x1)

## 2. Line Intersection
Two line segments intersect if and only if they have different orientations relative to each other.

## 3. Convex Hull
The smallest convex polygon that contains all given points.
- Graham Scan: O(N log N)
- Jarvis March: O(NH)
