# Advanced Math for Competitive Programming

## 1. Combinatorics
- **nCr (Binomial Coefficient)**: Number of ways to choose $r$ objects from $n$.
- **Pascal's Triangle**: Used to precompute nCr values in $O(N^2)$.

## 2. Modular Inverse
The modular multiplicative inverse of $a$ modulo $m$ is $x$ such that $ax \equiv 1 \pmod m$.
- Found using Extended Euclidean Algorithm or Fermat's Little Theorem (if $m$ is prime).

## 3. Fermat's Little Theorem
If $p$ is a prime number, then for any integer $a$, $a^p \equiv a \pmod p$.
- Specifically, if $a$ is not divisible by $p$, $a^{p-1} \equiv 1 \pmod p$.
- Useful for finding modular inverse: $a^{-1} \equiv a^{p-2} \pmod p$.

## 4. Matrix Exponentiation
A technique to find the $n$-th term of a linear recurrence (like Fibonacci) in $O(k^3 \log n)$ time, where $k$ is the size of the transition matrix.

## 5. Probability
- **Expected Value**: $E[X] = \sum x \cdot P(X=x)$.
- **Linearity of Expectation**: $E[X+Y] = E[X] + E[Y]$, even if $X$ and $Y$ are dependent.
