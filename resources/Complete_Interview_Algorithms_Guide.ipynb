# Complete Interview Algorithms Guide

This notebook explains the algorithms and patterns that are most frequently asked in coding interviews in very simple language.

The goal is not just to memorize names. The goal is to know:
- what problem the algorithm solves
- when to think about it
- how to explain it in normal words
- the standard approach
- time and space complexity
- the mistakes people usually make

---

## Golden rule

When you see a problem, do not ask only "How do I code this?"

First ask:
1. What type of problem is this?
2. What is the brute-force way?
3. What is the bottleneck?
4. Which known pattern removes that bottleneck?

## 1. Arrays and Hashing

Many interview problems are secretly array problems even when the story sounds fancy.

| Algorithm / Pattern | Layman terms | When to think of it | Complexity |
|---|---|---|---|
| Linear Scan | Look at every item once and keep updating the answer. | max, min, count, existence checks | O(n) |
| Prefix Sum | Store running totals so range sums become fast. | many subarray sum / range sum questions | O(n) build, O(1) query |
| Difference Array | Mark where updates start and stop, then rebuild once. | many range increment updates | O(n + q) |
| Hash Map | Store values so finding/counting becomes fast. | frequency, duplicates, pair sum | O(n) average |
| Hash Set | Store only presence. | repeated elements, seen-before logic | O(n) average |
| Kadane's Algorithm | Keep the best subarray ending at this position. | maximum subarray sum | O(n) |

### Prefix Sum in simple words

Instead of finding the sum from scratch again and again, we do some work once in the beginning.
Then each future query becomes easy.

```text
prefix[0] = 0
prefix[i] = sum of first i elements
sum from l to r = prefix[r + 1] - prefix[l]
```

### Hashing in simple words

A hash map is like a very fast notebook where you write:
- how many times something appeared
- where you saw it first
- whether it already exists

### Common interview signals

- "find duplicates"
- "how many times"
- "pair with target sum"
- "subarray sum equals k"
- "longest consecutive sequence"

### Common mistakes

- off-by-one errors in prefix sum
- forgetting negative numbers can break sliding assumptions but not hash-map counting
- using nested loops when a frequency table would do

## 2. Two Pointers and Sliding Window

These are two of the most asked patterns in interviews.

| Pattern | Layman terms | Interview signal | Complexity |
|---|---|---|---|
| Two Pointers | Use two indexes instead of trying every pair. | sorted array, pair sum, remove duplicates | O(n) or O(n log n) with sorting |
| Sliding Window | Grow and shrink a valid range while scanning once. | substring/subarray, longest/shortest, at most k | O(n) |

### Two Pointers in simple words

Imagine two fingers moving on the array.
If the array is sorted, you can decide intelligently which finger should move.
That is why you avoid checking every pair.

### Sliding Window in simple words

Keep a window that represents the current valid answer range.
Move the right pointer to include more things.
If the window becomes invalid, move the left pointer until it becomes valid again.

```text
left = 0
for right in range(n):
    include arr[right]
    while window is invalid:
        remove arr[left]
        left += 1
    update answer
```

### Common interview signals

- longest substring without repeating characters
- minimum window substring
- at most k distinct characters
- smallest subarray with sum at least target

### Common mistakes

- forgetting to shrink the window fully
- using sliding window when negative numbers break the monotonic behavior
- trying two pointers on unsorted data without first checking whether sorting is allowed

## 3. Binary Search

Binary search is not only for finding a number in a sorted array.
It is also used to search on the answer.

| Variant | Layman terms | Use case | Complexity |
|---|---|---|---|
| Classic Binary Search | Search in sorted data by cutting half each step. | exact value search | O(log n) |
| Lower Bound / Upper Bound | Find first or last valid position. | boundaries, insertion position | O(log n) |
| Binary Search on Answer | Guess an answer and test if it is possible. | minimum possible / maximum possible questions | O(log range * check_cost) |

### Layman explanation

If the answer space has a monotonic property, binary search works.
Monotonic means: once something becomes true, it stays true after that point.

Example:
- Can we finish in 5 days? No.
- Can we finish in 6 days? No.
- Can we finish in 7 days? Yes.
- Can we finish in 8 days? Yes.

This true/false border is perfect for binary search.

### Common interview signals

- sorted array
- find first / find last
- minimum possible value
- maximum possible value
- rotated sorted array

### Common mistakes

- wrong mid formula or infinite loop
- not deciding whether mid belongs to left or right half
- forgetting to keep the best valid answer in binary search on answer

## 4. Sorting, Intervals, and Greedy

Sorting often reveals the hidden structure of the problem.

| Algorithm / Pattern | Layman terms | Use case | Complexity |
|---|---|---|---|
| Merge Sort | Split, sort, and merge. | stable sorting, inversion count | O(n log n) |
| Quick Sort | Pick a pivot and partition around it. | general sorting | O(n log n) average |
| Merge Intervals | Sort intervals, then combine overlaps. | meetings, schedule, ranges | O(n log n) |
| Activity Selection Greedy | Always choose the interval that ends earliest. | maximum non-overlapping intervals | O(n log n) |
| Quickselect | Partition until kth element is found. | kth smallest/largest | O(n) average |
| Heap for Top K | Keep only the most important k items. | top k frequent, kth largest stream | O(n log k) |

### Greedy in simple words

Greedy means making the best-looking choice right now.
It works only when that local choice still leads to a global best answer.

### How to justify greedy in an interview

Do not just say "I used greedy".
Say:
- why the greedy choice is safe
- why any optimal solution can be transformed to use that choice

### Common interview signals

- intervals
- scheduling
- maximum number of non-overlapping tasks
- merge overlapping ranges
- kth largest / top k

### Common mistakes

- sorting by the wrong field
- assuming greedy works without proof
- fully sorting when heap or quickselect is enough

## 5. Linked List Algorithms

Linked list questions mostly test pointer control and edge-case discipline.

| Algorithm / Trick | Layman terms | Use case | Complexity |
|---|---|---|---|
| Traversal | Follow the next pointers one by one. | search, count, print | O(n) |
| Reverse Linked List | Flip arrows one by one. | reversal, reorder tasks | O(n) |
| Fast and Slow Pointers | One pointer moves faster than the other. | cycle, middle node, palindrome | O(n) |
| Merge Two Sorted Lists | Always take the smaller current node. | sorted merge | O(n + m) |
| Dummy Node | Add a fake starting node to simplify edge cases. | insert/delete near head | O(1) extra |

### Layman explanation

A linked list is like a chain of train coaches.
You cannot jump directly to coach 20.
You must keep following the link from one coach to the next.

### Standard reverse idea

```text
prev = null
curr = head
while curr != null:
    next = curr.next
    curr.next = prev
    prev = curr
    curr = next
return prev
```

### Common mistakes

- losing the next node while reversing
- forgetting head changes after reversal
- handling delete/insert around the first node badly

## 6. Stack, Queue, Heap, and Monotonic Structures

These data structures are very common in interview problems.

| Structure / Pattern | Layman terms | Interview signal | Complexity |
|---|---|---|---|
| Stack | Last in, first out. | brackets, undo, recursion-like behavior | O(1) push/pop |
| Queue | First in, first out. | BFS, level processing | O(1) amortized |
| Deque | Add/remove from both ends. | sliding window max/min | O(n) total in many problems |
| Heap / Priority Queue | Quickly get smallest or largest item. | top k, shortest path, best candidate | O(log n) updates |
| Monotonic Stack | Keep stack in increasing/decreasing order. | next greater/smaller, histogram | O(n) |
| Monotonic Deque | Keep useful candidates only. | sliding window max/min | O(n) |

### Layman explanation

- Stack: pile of plates
- Queue: line at a counter
- Heap: a special box where the best item is easiest to remove

### Why monotonic stack is powerful

When a new element comes, some old elements become useless forever.
We remove them immediately. That is why we turn many O(n^2) scans into O(n).

### Common interview signals

- next greater element
- largest rectangle in histogram
- daily temperatures
- sliding window maximum
- kth largest element

### Common mistakes

- storing values when indexes are actually needed
- forgetting to process remaining stack items
- mixing min-heap and max-heap logic

## 7. Trees and BSTs

Tree problems become easier when you decide what each recursive call should return.

| Algorithm / Pattern | Layman terms | Use case | Complexity |
|---|---|---|---|
| DFS Traversal | Go deep before coming back. | subtree answers, validation | O(n) |
| BFS Level Order | Visit nodes level by level. | shortest depth, level views | O(n) |
| Inorder Traversal | left -> node -> right | BST sorted order | O(n) |
| Preorder Traversal | node -> left -> right | serialization, cloning | O(n) |
| Postorder Traversal | left -> right -> node | bottom-up answers | O(n) |
| Lowest Common Ancestor | Find deepest shared parent of two nodes. | ancestor relationship | O(n) basic |
| BST Search / Insert | Use ordering to move left or right. | BST operations | O(h) |
| Tree Diameter | Longest path in the tree. | longest path problems | O(n) |

### Layman explanation

A tree is a family hierarchy.
Many questions become simple once you ask:
"What answer should one subtree return to its parent?"

### Common interview signals

- root, left, right
- level order
- path sum
- validate BST
- kth smallest in BST
- ancestor problems

### Common mistakes

- no base case for null node
- mixing local answer and global answer
- not understanding what recursion returns

## 8. Graph Algorithms

Many real interview stories are graph problems in disguise.

| Algorithm | Layman terms | Interview signal | Complexity |
|---|---|---|---|
| BFS | Spread one level at a time. | shortest path in unweighted graph | O(V + E) |
| DFS | Explore one path fully first. | components, cycles, reachability | O(V + E) |
| Topological Sort | Order tasks so prerequisites come first. | course schedule, dependency graph | O(V + E) |
| Union-Find / DSU | Quickly merge groups and check connectivity. | allowed swaps, connected groups | nearly O(1) amortized |
| Dijkstra | Always expand the cheapest known path next. | weighted shortest path, no negative edges | O((V + E) log V) |
| Bellman-Ford | Relax every edge again and again. | negative edges | O(VE) |
| Floyd-Warshall | Try every node as a middle stop between every pair. | all-pairs shortest path | O(V^3) |
| Kruskal | Sort edges and add safe edges only. | minimum spanning tree | O(E log E) |
| Prim | Grow the cheapest spanning tree from a node. | minimum spanning tree | O(E log V) |

### Layman explanation

- BFS is like water spreading level by level.
- DFS is like exploring one tunnel until the end.
- Union-Find is like quickly grouping friends into clubs.
- Dijkstra is like always continuing from the cheapest route discovered so far.

### Common interview signals

- connected components
- number of islands
- minimum steps
- cheapest route
- prerequisites / ordering
- can swap if connected

### Common mistakes

- using BFS on weighted shortest path
- forgetting visited checks
- using Dijkstra when negative edges exist

## 9. Recursion and Backtracking

Backtracking is used when you must explore all valid choices.

| Pattern | Layman terms | Use case | Complexity |
|---|---|---|---|
| Basic Recursion | Solve smaller version of the same problem. | trees, math, divide-and-conquer | varies |
| Subsets | For each item, choose take or skip. | power set | O(2^n) |
| Permutations | Try every possible order. | ordering problems | O(n * n!) |
| Combination Sum | Keep building until target is reached or broken. | target sum combinations | exponential |
| N-Queens | Place pieces without conflict. | board constraints | exponential |
| Sudoku Solver | Fill empty spots with valid choices only. | constraint solving | exponential |

### Backtracking in simple words

Choose.
Go deeper.
If it fails, undo that choice.
Try the next one.

```text
backtrack(state):
    if state is complete:
        save answer
        return

    for every valid choice:
        make choice
        backtrack(new state)
        undo choice
```

### Common interview signals

- all combinations
- all permutations
- generate all valid arrangements
- choose numbers / letters under constraints

### Common mistakes

- forgetting to undo the choice
- generating duplicates with repeated values
- not pruning obviously bad branches

## 10. Dynamic Programming

Dynamic Programming is about storing repeated work so we do not solve the same smaller problem again and again.

| DP Pattern | Layman terms | Use case | Complexity |
|---|---|---|---|
| 1D DP | Current answer depends on a few previous answers. | climbing stairs, house robber | O(n) |
| 2D / Grid DP | Cell answer depends on neighbors. | unique paths, min path sum | O(rows * cols) |
| Knapsack DP | For each item, decide take or skip. | capacity/value problems | O(n * capacity) |
| LCS / Edit Distance | Compare prefixes of two strings. | string similarity | O(nm) |
| LIS | Best increasing subsequence so far. | sequence order optimization | O(n log n) best |
| Interval DP | Solve smaller intervals first. | burst balloons, palindromes | O(n^3) often |
| Bitmask DP | Use bits to represent chosen set. | assignment, TSP-style | O(n * 2^n) |
| State Machine DP | Track modes like hold, sell, rest. | stock problems | O(n * states) |

### DP in simple words

If a problem keeps asking for answers to the same smaller pieces, save those answers.
That is DP.

### The most important DP checklist

1. What does dp[i] or dp[i][j] mean?
2. What are the base cases?
3. How do we move from smaller states to bigger states?
4. In what order should the table be filled?
5. Can space be reduced?

### Common interview signals

- number of ways
- maximum / minimum value
- choose or skip
- overlapping subproblems
- grid movement
- compare two strings or sequences

### Common mistakes

- writing loops before defining the state clearly
- wrong base cases
- filling the table in the wrong order

## 11. String Algorithms

Some string problems are simple scans. Others need specialized algorithms.

| Algorithm / Pattern | Layman terms | Use case | Complexity |
|---|---|---|---|
| Frequency Counting | Count characters quickly. | anagrams, palindrome checks | O(n) |
| Sliding Window | Keep a valid substring range. | longest unique substring, min window | O(n) |
| Trie | Store words as paths of characters. | prefix search, autocomplete | O(length) per operation |
| KMP | Use prefix knowledge so matching does not restart from zero. | pattern search | O(n + m) |
| Rabin-Karp | Compare rolling hashes of substrings. | pattern matching | O(n + m) average |
| Z Algorithm | Find how much each suffix matches the prefix. | string matching, prefix analysis | O(n) |
| Manacher's Algorithm | Reuse palindrome information around centers. | longest palindromic substring | O(n) |

### Layman explanation

- Trie is like a family tree of letters.
- KMP remembers how much of the pattern already matched.
- Rolling hash gives each substring a changing fingerprint.

### Interview advice

Do not jump to KMP first for every string problem.
First check whether a simple frequency array, sliding window, or two-pointer scan is enough.

### Common mistakes

- bad duplicate handling in frequency maps
- forgetting uppercase/lowercase rules
- choosing advanced algorithms when simple linear methods work

## 12. Math, Number Theory, and Bit Manipulation

These problems often become easy once you know a few standard tricks.

| Algorithm / Trick | Layman terms | Use case | Complexity |
|---|---|---|---|
| Euclidean GCD | Keep replacing bigger problem with remainder problem. | gcd, lcm, divisibility | O(log n) |
| Sieve of Eratosthenes | Mark multiples to find all primes. | many prime queries | O(n log log n) |
| Prime Factorization | Break number into prime building blocks. | divisor and factor problems | O(sqrt n) basic |
| Modular Exponentiation | Square and reduce instead of multiplying slowly. | huge powers under modulo | O(log exponent) |
| Extended Euclid | GCD plus coefficients. | modular inverse | O(log n) |
| XOR Trick | Equal numbers cancel each other. | single number, parity tricks | O(n) |
| Brian Kernighan Bit Count | Remove one set bit at a time. | count set bits | O(number of set bits) |
| Bitmasking | Use bits like on/off switches. | subset enumeration, compact states | often O(2^n) |

### Useful bit facts

```text
x & (x - 1) removes the lowest set bit
if x > 0 and (x & (x - 1)) == 0, then x is a power of two
a ^ a = 0
a ^ 0 = a
```

### Common mistakes

- integer overflow
- modulo mistakes
- using floating point when integer math is safer

## 13. Advanced but Frequently Respected Algorithms

These are not always asked in basic interviews, but strong companies and harder rounds like them.

| Algorithm | Layman terms | Use case | Complexity |
|---|---|---|---|
| Segment Tree | Store range answers in a tree so updates and queries stay fast. | range sum/min/max with updates | O(log n) |
| Fenwick Tree / BIT | Lightweight structure for prefix sums with updates. | dynamic prefix/range sums | O(log n) |
| Sparse Table | Precompute static range answers. | immutable range min/max/gcd | O(1) query after O(n log n) build |
| Binary Lifting | Jump ancestors in powers of two. | kth ancestor, LCA | O(log n) |
| Tarjan SCC | One DFS to find strongly connected components. | directed graph structure | O(V + E) |
| Kosaraju | DFS order, reverse graph, DFS again. | strongly connected components | O(V + E) |
| Aho-Corasick | Search many patterns at the same time. | multi-pattern matching | near linear |
| Sweep Line | Process sorted events from left to right. | interval overlaps, geometry | often O(n log n) |

### When to study these

First master the fundamentals:
- arrays and hashing
- sliding window
- two pointers
- trees
- BFS / DFS
- heaps
- binary search
- DP basics

Then add these advanced tools.

## 14. Quick Recognition Cheat Sheet

| If you see this in a problem | Think of this first |
|---|---|
| contiguous subarray / substring | sliding window, prefix sum |
| sorted array | two pointers, binary search |
| top k / kth largest | heap, quickselect |
| all possible ways | backtracking, DP |
| minimum steps in unweighted moves | BFS |
| cheapest route with positive weights | Dijkstra |
| connected groups / swaps allowed | Union-Find |
| prerequisites / ordering | topological sort |
| range query + updates | segment tree, Fenwick tree |
| longest increasing | LIS, DP, binary search |
| next greater / smaller | monotonic stack |
| choose under capacity | knapsack DP |
| pattern in text | KMP, Rabin-Karp, Z algorithm |
| palindrome substring | expand around center, DP, Manacher |

### Strong interview explanation formula

Say your approach in this order:
1. Brute force would be ...
2. That is too slow because ...
3. The key observation is ...
4. So I will use ...
5. The invariant is ...
6. Complexity is ...

### Final reminder

The real goal is not to memorize definitions.
The real goal is to look at a new problem and say:
- I know what family this problem belongs to
- I know the standard tool for that family
- I can explain why my approach is correct
- I can implement it without breaking edge cases
